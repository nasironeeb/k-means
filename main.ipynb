#  Example : IRIS

```python
iris = pd.read_csv('iris_data.csv')
```

## Data pre-processing

```python
irisX_df=pd.DataFrame(iris, columns = ['Longueur de la Sépale(cm)',
                                       'Largeur de la Sépale(cm)',
                                       'Longueur de la Pétale(cm)',
                                       'Largeur de la Pétale(cm)'])
irisY_df=pd.DataFrame(iris, columns = ['Type'])
```

```python
label_id = np.array(irisY_df)
for i in range(len(irisY_df)):
    if list(irisY_df.iloc[i])[0] =='setosa' :
        label_id[i] = 0
    elif list(irisY_df.iloc[i])[0] =='versicolor' :
        label_id[i] = 1
    elif list(irisY_df.iloc[i])[0] =='virginica' :
        label_id[i] = 2
# plot the points with the true label
plt.scatter(irisX_df['Longueur de la Sépale(cm)'], irisX_df['Largeur de la Sépale(cm)'],
            c=label_id, s=50, edgecolor='k')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#data_input will be the list of points
data_input=[]
for i in range(len(irisX_df)):
    data_input.append(list(irisX_df.iloc[i]))
```

## Application of k-means

```python
clusters = k_means(3, data_input, True, distance_euclidean, 100)
```

## Plot the results

```python
dimensions = [0, 1, 2, 3]
colors = ['green', 'skyblue', 'yellow', 'black', 'purple', 'blue']
cluster_types = ['Setosa', 'Versicolor', 'Virginica']
for i, center in enumerate(clusters):
    x = [point[dimensions[0]] for point in clusters[center].points]
    y = [point[dimensions[1]] for point in clusters[center].points]
    x1 = [point[dimensions[2]] for point in clusters[center].points]
    y1 = [point[dimensions[3]] for point in clusters[center].points] 
    plt.scatter(x, y, c=colors[i], label=cluster_types[i])
    plt.scatter(clusters[center].coord[dimensions[0]], clusters[center].coord[dimensions[1]], c='red', marker='+', s=100)
plt.ylabel('Sepal Width')
plt.xlabel('Sepal Length')
plt.legend(loc=1)
plt.show()
```

```python
for i, center in enumerate(clusters):
    x = [point[dimensions[0]] for point in clusters[center].points]
    y = [point[dimensions[1]] for point in clusters[center].points]
    x1 = [point[dimensions[2]] for point in clusters[center].points]
    y1 = [point[dimensions[3]] for point in clusters[center].points] 
    plt.scatter(x1, x, c=colors[i], label=cluster_types[i])
    plt.scatter(clusters[center].coord[dimensions[2]], clusters[center].coord[dimensions[0]], c='red', marker='+', s=100)
plt.ylabel('Petal Width')
plt.xlabel('Sepale width')
plt.legend(loc=2)
plt.show()
```

# Example : Random Point

## Generate the data

```python
# define the number of points
n_points = 1000

# generate random x and y and z coordinates between 1 and 100
x = np.random.rand(n_points) * 99 + 1
y = np.random.rand(n_points) * 99 + 1
z = np.random.rand(n_points) * 99 + 1

#data_input will be thee list of points
data_random_input=[]
for i in range(n_points):
    data_random_input.append([x[i],y[i],z[i]])
```

## Plot the 2D plot

```python
plt.scatter(x,y)
plt.title('Set of 2D random points')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## Plot the 3D plot

```python

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter( x, y, z)
ax.set_title ('Set of 3D random points')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
clusters = k_means(4, data_random_input, False, distance_euclidean, 5000)
```
